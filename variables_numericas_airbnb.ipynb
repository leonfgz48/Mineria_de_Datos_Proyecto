```python
# DATOS CREADOS POR EMMANUEL

categoricas = pd.read_csv('../data/variables_categoricas_airbnb.csv')
categoricas['id'] = categoricas['id'].astype('int')
categoricas = categoricas.set_index('id')
```

```python
# DATOS ORIGINALES

listings = pd.read_csv('../data/raw_list.csv')
listings['id'] = listings['id'].astype('int')
listings = listings.set_index('id')
```

```python
listings.columns.sort_values()
```

```python
listings[['number_of_reviews']]
```

```python
num_cols = ['host_total_listings_count',
            'host_acceptance_rate',
            'host_response_rate',
            'calculated_host_listings_count_entire_homes',
            'calculated_host_listings_count_private_rooms',
            'calculated_host_listings_count_shared_rooms',
            'latitude',
            'longitude',
            'accommodates',
            'bedrooms',
            'beds',
            'price']
```

```python
listings_num = listings[num_cols]
```

```python
listings_num.corr()
```

# Limpieza de columnas en texto

```python
listings_num['host_acceptance_rate'] = listings_num['host_acceptance_rate'].str.replace('%', '')\
                                       .astype(float) / 100

listings_num['host_response_rate'] = listings_num['host_response_rate'].str.replace('%', '')\
                                     .astype(float) / 100
```

```python
listings_num['price'] = listings_num['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
listings_num
```

# Valores Nulos

```python
listings_num.isnull().sum()
```

```python
listings_num['bedrooms'].fillna(0, inplace=True)
listings_num['beds'].fillna(1, inplace=True)
```

```python
listings_num['host_acceptance_rate'].fillna(listings_num['host_acceptance_rate'].mean(), inplace=True)
listings_num['host_response_rate'].fillna(listings_num['host_response_rate'].mean(), inplace=True)
```

```python
listings_num = listings_num.dropna(subset=['price'])
```

```python
listings_num.isnull().sum()
```

# Manejo de Valores Atípicos

```python
Q1 = listings_num['host_total_listings_count'].quantile(0.25)
Q3 = listings_num['host_total_listings_count'].quantile(0.75)
IQR = Q3 - Q1
listings_num_clean = listings_num[~((listings_num['host_total_listings_count'] < (Q1 - 1.5 * IQR)) | 
                                    (listings_num['host_total_listings_count'] > (Q3 + 1.5 * IQR)))]
```

```python
quantile_99 = listings_num_clean['price'].quantile(0.999)

listings_num_clean = listings_num_clean[listings_num_clean['price'] <= quantile_99]
```

# Unión de dataframes con categóricos

```python
listings_num_clean
```

```python
listings_num_clean.index = listings_num_clean.index.astype(int)
categoricas.index = categoricas.index.astype(int)
```

```python
listings_processed_s_val_at = pd.merge(listings_num_clean, 
                              categoricas, 
                              left_index=True, 
                              right_index=True, 
                              how='inner')
```

```python
listings_processed_c_val_at = pd.merge(listings_num, 
                              categoricas, 
                              left_index=True, 
                              right_index=True, 
                              how='inner')
```
